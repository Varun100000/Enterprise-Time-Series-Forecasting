```python
close = pd.read_csv('sp500_close.csv', index_col = 0, parse_dates = True)
```

```python
goog = close[['GOOG']].copy().dropna()
```

```python
goog.head()
```

```python
goog.plot()
```

```python
# calculating log returns
goog_ret = np.log(goog.pct_change(1) + 1) 
#goog.pct_change(1) is just giving the net return which is [p(t+1) - P(t)]/p(t)
#taking log of (1 + net_return) is essentially log of gross returns , or otherwise called gross returns
# 1 + net_return = p(t+1)/p(t)
# log(1 + net_return) = log(p(t+1)/p(t)) =  log(p(t+1)) - log(p(t)) = log price at time t+1 - log price at time t
```

```python
goog.pct_change(1)
```

```python
goog_ret.plot();
```
