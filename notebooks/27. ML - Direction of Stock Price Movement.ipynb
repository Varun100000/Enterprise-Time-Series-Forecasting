```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
os.chdir(r'G:\My Drive\Study\Time Series Analysis Udemy\Data\time series analysis forecasting')
os.getcwd()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import r2_score, mean_absolute_percentage_error
```

```python
df0 = pd.read_csv('sp500sub.csv', index_col = 'Date', parse_dates = True)
```

```python
df = df0[df0['Name']=='IBM'][['Close']].copy()
```

```python
df['LogClose'] = np.log(df['Close'])
```

```python
df['LogReturn'] = df['LogClose'].diff()
```

```python
df.shape
```

```python
Ntest = 252
train = df.iloc[:-Ntest]
test = df.iloc[-Ntest:]
```

```python
series =  df['LogReturn'].to_numpy()[1:] #1st value is 0 in differencing # converting this to numpy array since it is easy to index
target = (series > 0) * 1 #multiplying it by 1 to convert true and false to 1 and 0

T = 21
X = []
Y = []

for t in range(len(series)-T):
    x = series[t:t+T] # returns from t to t + T -1
    y = target[t+T] # returns t+T #this is not coming from time series but a precomputed target variable
    X.append(x)
    Y.append(y)

X = np.array(X).reshape(-1, T)
Y = np.array(Y)
N = len(X)
print("X.shape", X.shape, "Y.shape", Y.shape)
```

```python
target
```

```python
Xtrain, Ytrain = X[:-Ntest], Y[:-Ntest]
Xtest, Ytest = X[-Ntest:], Y[-Ntest:] 
```

```python
lr = LogisticRegression()
lr.fit(Xtrain, Ytrain)
lr.score(Xtrain, Ytrain)
```

```python
'''accuracy is just 50% not great'''
```

```python
lr.score(Xtest, Ytest)
```

```python
''' this is just like predicting a coin toss'''
```

```python
svc =  SVC()
svc.fit(Xtrain, Ytrain)
svc.score(Xtrain, Ytrain)
```

```python
''' this does better on the train set'''
```

```python
svc.score(Xtest, Ytest)
```

```python
' model just overfits on the train set'''
```

```python
rf =  RandomForestClassifier()
rf.fit(Xtrain, Ytrain)
rf.score(Xtrain, Ytrain)
```

```python
''' this is just overfitting'''
```
