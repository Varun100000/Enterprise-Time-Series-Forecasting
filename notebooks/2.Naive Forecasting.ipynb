```python
df = pd.read_csv('SPY.csv', index_col ='Date', parse_dates=True )
```

```python
df.head()
```

```python
df['ClosePrediction'] = df['Close'].shift(1)
```

```python
df.head()
```

```python
y_true = df.iloc[1:]['Close']
y_pred = df.iloc[1:]['ClosePrediction']
```

## Metrics

```python
# SSE
(y_true - y_pred).dot(y_true - y_pred)
```

```python
# MSE = SSE/n
mean_squared_error(y_true, y_pred)
```

```python
# MSE calculating it manually
(y_true - y_pred).dot(y_true - y_pred) / len(y_true)
```

```python
#RMSE
# mean_squared_error(y_true, y_pred, squared = False) # no longer works in the newer version of sklearn
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(y_true, y_pred)
```

```python
np.mean(y_true)
```

```python
# looking at the root mean squared error and comparing this with the mean of y_true, it shows that RMSE as a percentage of mean of y_true is very small which is great
# and it is great because it is naive forecast
```

```python
#RMSE done manually
np.sqrt((y_true - y_pred).dot(y_true - y_pred) / len(y_true))
```

```python
#mean absolute error
mean_absolute_error(y_pred,y_true)
```

```python
#mean absolute error done manually
np.sum(abs(y_true-y_pred))/len(y_true)
```

```python
r2_score(y_true, y_pred)
## THIS LOOKS AWESOME SINCE MORE THAN 99% ERRORS EXPLAINABLE
```

```python
#MAPE
mean_absolute_percentage_error(y_true, y_pred)
```

```python
np.mean((abs(y_pred-y_true)/abs(y_true)))
```

```python
#sMAPE
numerator = abs(y_pred-y_true)
denominator = (abs(y_pred) + abs(y_true))/2
np.mean(numerator/denominator)
```
