```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
T = 1000 # num of time steps
p0 = 10 # initial price
mu = 0.001 # drift # this changes how we want the long term trend to look like - +ve or -ve # used to control the trend of the time series
```

```python
# initial log price
last_p = np.log(p0)
```

```python
log_returns = np.zeros(T)
prices = np.zeros(T)
```

```python
for t in range(T):
    # log return sampling from normal distribution at each time step
    r = 0.01 * np.random.randn()

    # computing new log price - using log return = log price at t minus log price at t-1
    p = r + mu + last_p

    # storing the returns and log prices
    log_returns[t] = r + mu
    prices[t] = np.exp(p)

    last_p = p
```

```python
plt.figure(figsize=(20,8))
plt.plot(prices)
plt.title("Simulated prices")
```

```python
plt.figure(figsize = (20,8))
plt.plot(log_returns)
plt.title("random returns plus drift")
```

### vectorized version of the above loop

```python
prices = np.zeros(T)
log_returns = np.zeros(T)

log_returns = 0.01 * np.random.randn(T) + mu

log_prices = np.log(p0) + np.cumsum(log_returns)

prices  = np.exp(log_prices)
```

```python
plt.figure(figsize=(20,8))
plt.plot(prices)
plt.title("Simulated prices")
```

```python
plt.figure(figsize = (20,8))
plt.plot(log_returns)
plt.title("random returns plus drift")
```
